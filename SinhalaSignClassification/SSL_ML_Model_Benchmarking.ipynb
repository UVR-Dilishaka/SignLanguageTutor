```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import seaborn as sns

# Example of true labels and predicted values (Replace with actual predictions)
true_labels = np.array([0, 1, 2, 0, 1, 2, 0, 1, 2, 1])  # Example true labels (adjust as needed)
predictions = {
    'PCA + Random Forest': np.array([0, 1, 2, 0, 1, 1, 0, 2, 2, 1]),
    'PCA + KNN': np.array([0, 1, 2, 0, 1, 2, 0, 1, 2, 2]),
    'PCA + Logistic Regression': np.array([0, 1, 1, 0, 1, 2, 0, 1, 2, 1]),
    'PCA + SVM': np.array([0, 1, 2, 0, 1, 1, 0, 1, 2, 1]),
    'LDA + Random Forest': np.array([0, 1, 2, 0, 1, 2, 0, 1, 1, 2]),
    'LDA + KNN': np.array([0, 1, 2, 0, 1, 2, 0, 2, 1, 1]),
    'LDA + Logistic Regression': np.array([0, 1, 2, 0, 1, 1, 0, 2, 2, 1]),
    'LDA + SVM': np.array([0, 1, 2, 0, 1, 1, 0, 1, 2, 1]),
}

# Create a DataFrame for the evaluation metrics
evaluation_metrics = []

for model_name, preds in predictions.items():
    # Compute confusion matrix
    cm = confusion_matrix(true_labels, preds)

    # Generate classification report (Precision, Recall, F1-Score for each class)
    class_report = classification_report(true_labels, preds, output_dict=True)

    # Collect overall precision, recall, and f1-score
    precision = precision_score(true_labels, preds, average='weighted', zero_division=0)
    recall = recall_score(true_labels, preds, average='weighted', zero_division=0)
    f1 = f1_score(true_labels, preds, average='weighted', zero_division=0)

    # Store evaluation metrics in a dictionary
    evaluation_metrics.append({
        'Model': model_name,
        'Accuracy': (preds == true_labels).mean(),
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix': cm,
        'Classification Report': class_report
    })

# Convert evaluation metrics to a DataFrame
metrics_df = pd.DataFrame(evaluation_metrics)

# Display the evaluation metrics
print(metrics_df[['Model', 'Accuracy', 'Precision', 'Recall', 'F1-Score']])

# Display confusion matrices for each model
for model in metrics_df['Model']:
    cm = metrics_df.loc[metrics_df['Model'] == model, 'Confusion Matrix'].values[0]
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'],
                yticklabels=['Class 0', 'Class 1', 'Class 2'])
    plt.title(f'Confusion Matrix for {model}')
    plt.ylabel('True Labels')
    plt.xlabel('Predicted Labels')
    plt.show()

# Show the classification reports for each model
for model in metrics_df['Model']:
    report = metrics_df.loc[metrics_df['Model'] == model, 'Classification Report'].values[0]
    print(f"\nClassification Report for {model}:\n")
    print(pd.DataFrame(report).transpose())
```
