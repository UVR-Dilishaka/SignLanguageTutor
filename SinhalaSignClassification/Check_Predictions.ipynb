# Check for all

```python
import pandas as pd
import joblib
import numpy as np

# Load the trained RandomForest model
model_path = r"/content/best_rf_model.pkl"
best_rf_model = joblib.load(model_path)

# Load the dataset
csv_path = r"/content/SSL.csv"
df = pd.read_csv(csv_path)

# Assuming the dataset has a 'label' column and 16 feature columns
feature_columns = df.columns[:-1]  # Assuming last column is 'label'
label_column = df.columns[-1]

# Get feature values and actual labels for all rows
X_test = df[feature_columns].values
y_actual = df[label_column].values

# Predict labels
y_pred = best_rf_model.predict(X_test)

# Compare predictions with actual labels
correct_count = sum(y_actual == y_pred)
total_samples = len(y_actual)
accuracy = correct_count / total_samples

# Print results with row numbers
print("Row Number | Expected Label | Predicted Label")
print("---------------------------------------------")
for i, (actual, predicted) in enumerate(zip(y_actual, y_pred), start=1):
    print(f"{i:<11} | {actual:<14} | {predicted}")

print(f"\nModel Accuracy on Full Dataset: {accuracy:.4f} ({correct_count}/{total_samples} correct)")
```

# Wrong predictions

```python
import pandas as pd
import joblib
import numpy as np

# Load the trained RandomForest model
model_path = r"/content/best_rf_model.pkl"
best_rf_model = joblib.load(model_path)

# Load the dataset
csv_path = r"/content/SSL.csv"
df = pd.read_csv(csv_path)

# Assuming the dataset has a 'label' column and 16 feature columns
feature_columns = df.columns[:-1]  # Assuming last column is 'label'
label_column = df.columns[-1]

# Get feature values and actual labels for all rows
X_test = df[feature_columns].values
y_actual = df[label_column].values

# Predict labels
y_pred = best_rf_model.predict(X_test)

# Identify incorrect predictions
incorrect_indices = np.where(y_actual != y_pred)[0]
incorrect_count = len(incorrect_indices)
total_samples = len(y_actual)

# Print incorrect predictions
print("Row Number | Expected Label | Predicted Label")
print("---------------------------------------------")
for i in incorrect_indices:
    print(f"{i+1:<11} | {y_actual[i]:<14} | {y_pred[i]}")  # +1 to match row number

# Print misclassification rate
misclassification_rate = incorrect_count / total_samples
print(f"\nTotal Incorrect Predictions: {incorrect_count}/{total_samples}")
print(f"Misclassification Rate: {misclassification_rate:.4f}")
```
