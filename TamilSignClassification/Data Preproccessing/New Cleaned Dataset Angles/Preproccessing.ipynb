## Combined the seperated CSV file into one ##

```python
# Dictionary mapping filenames to gesture labels
import pandas as pd


file_paths = [
    "New Cleaned Dataset\clean_tsl1.csv",
    "New Cleaned Dataset\clean_tsl2.csv",
    "New Cleaned Dataset\clean_tsl3.csv",
    "New Cleaned Dataset\clean_tsl4.csv",
    "New Cleaned Dataset\clean_tsl5.csv",
    "New Cleaned Dataset\clean_tsl6.csv",
    "New Cleaned Dataset\clean_tsl7.csv",
    "New Cleaned Dataset\clean_tsl8.csv",
    "New Cleaned Dataset\clean_tsl9.csv",
    "New Cleaned Dataset\clean_tsl10.csv",
    "New Cleaned Dataset\clean_tsl11.csv",
    "New Cleaned Dataset\clean_tsl12.csv",
]

gesture_labels = {
    "clean_tsl1.csv": 1,
    "clean_tsl2.csv": 2,
    "clean_tsl3.csv": 3,
    "clean_tsl4.csv": 4,
    "clean_tsl5.csv": 5,
    "clean_tsl6.csv": 6,
    "clean_tsl7.csv": 7,
    "clean_tsl8.csv": 8,
    "clean_tsl9.csv": 9,
    "clean_tsl10.csv": 10,
    "clean_tsl11.csv": 11,
    "clean_tsl12.csv": 12
}


# Initialize an empty list to store DataFrames
dataframes = []

# Iterate over the files, read them, and add a 'target' column with the gesture label
for file in file_paths:
    df = pd.read_csv(file)
    gesture_label = gesture_labels.get(file.split("\\")[-1], None)  # Get the gesture label from the dictionary
    df['target'] = gesture_label  # Add the gesture label as the 'target' column
    dataframes.append(df)

# Concatenate all DataFrames into one
combined_data = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a new CSV file
output_path = "Combined_With_Target.csv"
combined_data.to_csv(output_path, index=False)

print(f"Combined CSV with target column saved to: {output_path}")
```

## Removing the imgae column ##

```python
import pandas as pd

# Function to check and remove rows with null values in a CSV file and drop the 'image' column
def check_and_remove_null_values(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Check for null values in the DataFrame
    null_values = df.isnull().sum()
    
    print(f"Null values in each column of {file_path}:")
    print(null_values)
    
    # Remove rows with any null values
    df_cleaned = df.dropna()

    # Drop the 'image' column if it exists
    if 'Image' in df_cleaned.columns:
        df_cleaned = df_cleaned.drop(columns=['Image'])
        print("'Image' column dropped.")
    
    # Save the cleaned DataFrame to a new CSV file
    cleaned_file_path = file_path.replace(".csv", "cleaned.csv")
    df_cleaned.to_csv(cleaned_file_path, index=False)
    
    print(f"Cleaned CSV saved to: {cleaned_file_path}")
    return df_cleaned

# Example usage
# Call the function for the combined CSV file
combined_csv_path = "Combined_With_Target.csv"
check_and_remove_null_values(combined_csv_path)
```
