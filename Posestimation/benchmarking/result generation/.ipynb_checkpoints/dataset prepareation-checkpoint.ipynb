```python
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
df1 = pd.read_csv('mp_benchmarking_dataset.csv')
df1
```

```python
for i in range(0,21):
    df1[str(i)] = list(zip(df1[str(i) + '_x' ], df1[str(i) + '_y']))
```

```python
df1
```

```python
df1.to_csv('mp_final_dataset.csv', index=False)
```
